##**Confidence Interval**


```python
import numpy as np
import pandas as pd
import scipy.stats as stats
```

```python
data = np.array([
    1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29

])
```

```python
data
```

```python
n= len(data)
```

```python
print("sample size(n):",n)
```

```python
mean=np.mean(data)
std_sample=np.std(data, ddof=1)   # sample standard deviation
print("sample_mean:",round(mean,4))
print("sample standard ddeviation:",round(std_sample,4))
```

### Task A

```python
confidence_level=0.99
alpha = 1-confidence_level
```

```python
#critical value (t)
t_critical = stats.t.ppf(1-alpha/2, df=n-1)
```

```python
t_critical
```

```python
margin_of_error_t = t_critical*(std_sample/np.sqrt(n))   # margin of error in T
```

```python
margin_of_error_t
```

```python

# Confidence Interval
lower_t= mean - margin_of_error_t
upper_t= mean + margin_of_error_t
```

```python
print("\n--- 99% Confidence Interval (t-distribution) ---")
print("t-critical value:", round(t_critical, 4))
print("margin of Error:", round(margin_of_error_t, 4))
print("Confidence Interval: ({:.4f}, {:.4f})".format(lower_t, upper_t))
```

###Task B

```python
sigma = 0.2
# given population std
```

```python
# critical_value(z)
z_critical = stats.norm.ppf(1-alpha/2)
```

```python
z_critical
```

```python
# margin of error
margin_of_error_z = z_critical* (sigma/np.sqrt(n))
margin_of_error_z
```

```python
#confidence_interval
lower_z =mean-margin_of_error_z
upper_z =mean+margin_of_error_z
```

```python
print("\n--- 99% Confidence Interval (z-distribution) ---")
print("z-critical value:", round(z_critical, 4))
print("margin of Error:", round(margin_of_error_z, 4))
print("Confidence Interval: ({:.4f}, {:.4f})".format(lower_z, upper_z))
```

## **Hypothesis Testing**

```python
# weekly operating cost analysis
import math
from scipy.stats import norm
```

```python
#given info
x_bar = 3050    # Sample mean cost
n = 25    # Number of restaurants
X_mean = 600    # Mean units produced
X_sd = 25   # Standard deviation of units
```

```python
# Cost model: W = 1000 + 5X
mu = 1000 + 5 * X_mean   # Theoretical mean cost
sigma = 5 * X_sd    # Standard deviation of cost
```

```python
alpha = 0.05
```

```python
print("STEP 1: HYPOTHESES")
print("H0: Actual cost = Theoretical cost")
print("H1: Actual cost > Theoretical cost\n")
```

```python
z = (x_bar - mu) / (sigma / math.sqrt(n))
print("STEP 2: CALCULATE Z VALUE")
print("Theoretical Mean =", mu)
print("Standard Deviation =", sigma)
print("Z value =", round(z, 2), "\n")
```

```python
#critical value and p value
critical_value = norm.ppf(1 - alpha)
p_value = 1 - norm.cdf(z)
print("STEP 3: COMPARE VALUES")
print("Critical Value =", round(critical_value, 2))
print("P-value =", round(p_value, 5), "\n")
```

```python
if z > critical_value:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

```python
z
```

```python
print("\nSTEP 5: CONCLUSION")
if z > critical_value:
    print("Costs are higher than the model (Claim is TRUE)")
else:
    print("Costs are NOT higher than the model (Claim is FALSE)")
```
